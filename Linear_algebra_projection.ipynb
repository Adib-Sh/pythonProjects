```python
from numpy import *
from matplotlib.pyplot import *
from numpy.linalg import *
from scipy import optimize
import sys
from sympy import Symbol
from sympy.solvers import solve

def g(x,y,z):
    return 2*(x**2) - (y**2) + 2*(z**2) - 10*x*y - 4*x*z + 10*y*z -1

xvalues = linspace(-1,1,10)
yvalues = linspace(-1,1,10)

xline = []
yline = []
zline = []

for i in xvalues:
    for j in yvalues:
        z = Symbol('z')
        l = solve(g(i,j,z),z)
        xline.extend([i,i])
        yline.extend([j,j])
        z_1, z_2 = float(l[0]), float(l[1])
        zline.extend([z_1,z_2])



fig = figure()
ax = axes(projection='3d')
ax.scatter(xline,yline,zline)
ax.view_init(0,45)
```
